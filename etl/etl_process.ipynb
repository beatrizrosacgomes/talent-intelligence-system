Tratamento Bases Jobs

```python
import json
import pandas as pd
import re

# Função para limpar caracteres inválidos do Excel e quebras de linha
def limpar_texto(texto):
    if isinstance(texto, str):
        texto = texto.replace('\n', ' ').replace('\r', ' ')
        texto = re.sub(r'[\x00-\x1F\x7F-\x9F]', '', texto)  # Remove caracteres de controle
        texto = texto.replace('·', '-')  # Substitui pontos por hífen, se necessário
    return texto

# Caminho do arquivo JSON
caminho_arquivo = 'C:/Users/beatr/OneDrive/Documentos/projeto datathon/src/data/raw/jobs.json'

# Lendo o JSON
with open(caminho_arquivo, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Processar todos os IDs
registros = []

for id_vaga, conteudo in data.items():
    registro = {'id_vaga': id_vaga}
    for secao, campos in conteudo.items():
        for chave, valor in campos.items():
            nome_coluna = f"{secao}_{chave}"
            registro[nome_coluna] = limpar_texto(valor)
    registros.append(registro)

# Criar o DataFrame
df = pd.DataFrame(registros)

# Caminho de saída
output_path = 'C:/Users/beatr/OneDrive/Documentos/projeto datathon/src/data/silver/jobs.xlsx'

# Exportar para Excel
df.to_excel(output_path, index=False)

print(f"Arquivo Excel salvo com todos os dados em: {output_path}")
```

ETL - Applicants

```python
import json
import pandas as pd
import re
import unicodedata

# Função para limpar texto
def limpar_texto(texto):
    if not isinstance(texto, str):
        return texto

    # Remove caracteres de controle inválidos para XML (exceto \n, \r, \t)
    texto = re.sub(r'[\x00-\x08\x0B\x0C\x0E-\x1F\x7F-\x9F]', '', texto)

    # Remove caracteres não imprimíveis ou que não são considerados "legais" em XML
    texto = ''.join(c for c in texto if (
        c == '\t' or c == '\n' or c == '\r' or
        (0x20 <= ord(c) <= 0xD7FF) or
        (0xE000 <= ord(c) <= 0xFFFD)
    ))

    return texto


# Caminho do arquivo JSON
caminho_json = 'C:/Users/beatr/OneDrive/Documentos/projeto datathon/src/data/raw/applicants.json'

# Abrir o arquivo JSON
with open(caminho_json, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Lista para armazenar os registros tratados
registros = []

# Processar cada candidato
for id_candidato, conteudo in data.items():
    registro = {'id_candidato': id_candidato}
    for secao, campos in conteudo.items():
        if isinstance(campos, dict):
            for chave, valor in campos.items():
                nome_coluna = f"{secao}_{chave}"
                registro[nome_coluna] = limpar_texto(valor)
        else:
            # Trata campos que não são dicionários, como cv_pt ou cv_en
            registro[secao] = limpar_texto(campos)
    registros.append(registro)

# Criar DataFrame
df = pd.DataFrame(registros)

# Caminho de saída
caminho_saida = 'C:/Users/beatr/OneDrive/Documentos/projeto datathon/src/data/silver/applicants_tratado.xlsx'

# Exportar para Excel
df.to_excel(caminho_saida, index=False)

print(f"Arquivo Excel salvo com sucesso em: {caminho_saida}")
```

```python
import json
import pandas as pd

# Caminho para o arquivo JSON
caminho_arquivo = 'C:/Users/beatr/OneDrive/Documentos/projeto datathon/src/data/raw/prospects.json'

# Carregar o JSON
with open(caminho_arquivo, 'r', encoding='utf-8') as f:
    data = json.load(f)

# Lista para armazenar os dados tratados
prospects_tratados = []

# Iterar pelas vagas e prospectos
for codigo_vaga, dados_vaga in data.items():
    titulo = dados_vaga.get('titulo', '')
    modalidade = dados_vaga.get('modalidade', '')
    for prospect in dados_vaga.get('prospects', []):
        linha = {
            'codigo_vaga': codigo_vaga,
            'titulo': titulo,
            'modalidade': modalidade,
            'nome': prospect.get('nome', ''),
            'codigo': prospect.get('codigo', ''),
            'situacao_candidato': prospect.get('situacao_candidado', ''),
            'data_candidatura': prospect.get('data_candidatura', ''),
            'ultima_atualizacao': prospect.get('ultima_atualizacao', ''),
            'comentario': prospect.get('comentario', ''),
            'recrutador': prospect.get('recrutador', '')
        }
        prospects_tratados.append(linha)

# Criar o DataFrame
df = pd.DataFrame(prospects_tratados)

# Salvar como Excel
df.to_excel('C:/Users/beatr/OneDrive/Documentos/projeto datathon/src/data/silver/prospects_tratado.xlsx', index=False)

print("Arquivo salvo com sucesso em: src/data/raw/prospects_tratado.xlsx")
```
